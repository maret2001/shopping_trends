# --------------------------------------------------------------------------------------------------
# 3. Items Buying Activities Frequently

```python
df.columns
```

## Highest

```python
f_b = df.groupby('Item Purchased')['Previous Purchases'].sum().sort_values(ascending=False).head(5).reset_index()
f_b
```

```python
fig = px.bar(
    df.groupby('Item Purchased')['Previous Purchases']
    .sum()
    .sort_values(ascending=False).head(5)
    .reset_index(name="Previous Purchases"),
    x="Previous Purchases",
    y="Item Purchased",
    color="Previous Purchases",
    title="Top 5 Highest Items Buying Frequently",
)
fig.show()
```

## Lowest

```python
f_b = df.groupby('Item Purchased')['Previous Purchases'].sum().sort_values(ascending=True).head(5).reset_index()
f_b
```

```python
fig = px.bar(
    df.groupby('Item Purchased')['Previous Purchases']
    .sum()
    .sort_values(ascending=True).head(5)
    .reset_index(name="Previous Purchases"),
    x="Previous Purchases",
    y="Item Purchased",
    color="Previous Purchases",
    title="Top 5 Lowest Items Buying Frequently",
)
fig.show()
```
